# Identificando valores vazios

```python
import pandas as pd
import numpy as np

df = pd.read_csv("clientes.csv")
display(df)

#se tiver algum dados que nÃ£o seja nÃºmero, subistitua por vazio
df["TotalGasto"] = pd.to_numeric(df["TotalGasto"], errors="coerce")

# info
print(df.info())
# df.isna().any() ou .sum
#Mostra quando se a valores vazios na tabela
# print(df.isna().any())
print(df.isna().sum())
```

# Valores vazios nÃ£o identificaados como vazio

```python
# identifica na hora de importar
# valores_vazios = ["?","-"]
# df = pd.read_csv("clientes.csv", na_values=valores_vazios)



# identifica com replace
#
df = df.replace({"?": np.nan, "-": np.nan})
display(df)
```

# Removendo valores vazios

```python
# joga fora o valor vazio
# dropna()

# 0 -> exclui linhas, 1 ->exclui colunas
# all -> quando todos os valores estiverem vazios
# any -> quando tiver algum um dado vazio
df = df.dropna(how='any', axis=0)
```

# Tratar valores vazios
- Preenchendo com valores fixo

```python
media = df["MesesComoCliente"].mean()
df["MesesComoCliente"] = df["MesesComoCliente"].fillna(media)
display(df)
```

- Preenchendo com anterior ou seguinte

```python
df["MesesComoCliente"] = df["MesesComoCliente"].fillna(method="ffill")
display(df)
```

- Interpolando

```python
df["MesesComoCliente"] = df["MesesComoCliente"].interpolate()
display(df)
```
